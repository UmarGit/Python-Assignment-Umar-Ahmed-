```python
# Quetion 1
print("Welcome to the calculator")
print("-----------MENU----------")
flag = False
while flag != True :
    print("What do you want to do ?")
    
    print("->> Press a for Addition")
    
    print("->> Press b for Substraction")
    
    print("->> Press c for Multiplication")
    
    print("->> Press d for Divsion")
    
    print("->> Press e for Power")
    
    print("->> Press q to Quit\n")
    
    choice=input("Enter your choice : ")
    
    if choice == 'a':
        
        print(">>>>>ADDITION<<<<<")
        
        num1=float(input("Enter the first number to add !"))
    
        num2=float(input("Enter the second number to add !"))
        
        print("The addition of {} and {} is : {}".format(num1,num2,num1+num2))
    elif choice == 'b':
        
        print(">>>>>SUBSTRACTION<<<<<")
        
        num1=float(input("Enter the first number to substract !"))
    
        num2=float(input("Enter the second number to substract !"))
        
        print("The substraction of {} and {} is : {}".format(num1,num2,num1-num2))
    elif choice == 'c':
        
        print(">>>>>MULTIPLICATION<<<<<")
        
        num1=float(input("Enter the first number to multiply !"))
    
        num2=float(input("Enter the second number to multiply !"))
        
        print("The multiplication of {} and {} is : {}".format(num1,num2,num1*num2))
    elif choice == 'd':
        
        print(">>>>>DIVISION<<<<<")
        
        num1=float(input("Enter the first number to divide !"))
    
        num2=float(input("Enter the second number to divide !"))
        
        print("The divsion of {} and {} is : {}".format(num1,num2,num1/num2))
    elif choice == 'e':
        
        print(">>>>>POWER<<<<<")
        
        num1=float(input("Enter the first number !"))
    
        num2=float(input("Enter the power to calculate !"))
        
        print("The power of {} and {} is : {}".format(num1,num2,num1**num2))
    elif choice == 'q':
        
        print(">>>>>QUIT<<<<<")
        
        print("Thank You for using our calculator")
        
        flag = True
        
        break
    
        
```

```python
# Question 3
dictionary={}

flag = False

while flag != True :
    print("-->> Press a to add values in the dictionary")
    
    print("-->> Press any key to quit adding values in the dictionary\n\n")
    
    choice = input("Enter your choice : ")
    
    if choice == 'a':
        key = input("Enter the Key to add ")
        
        value = input("Enter the Value to add ")
        
        dictionary[key] = value
    else:
        
        print("\nThe Dictionary is \n")
        
        print(dictionary)
        
        flag = True
        
        break
        
    
```

```python
# Question 4
dictionary = {
    "salary_1" : 100 ,
    "salary_2" : 200 ,
    300 : "salary_3" ,
    400 : "salary_4" ,
}

sum=0

for key_sum,salary_sum in dictionary.items() :
    
    if type(key_sum) == int :
        sum += key_sum
        
    elif type(salary_sum) == int :
        sum += salary_sum

print("The sum of all numeric values in a dictionary are : {}".format(sum))    


```
